# Life Expectancy and GDP


## Introduction

This project aims to analyze data that comes from both the WHO (World Health Organization) and the World Bank in an attempt to answer questions regarding GDP and life expectancy. The data in this project comes from the following six nations: China, Chile, Germany, Mexico, United States, and Zimbabwe. 


Some of the key points addressed in this project will include a look at the increase or decrease of GDP over time, Average GDP per country, Average life expectancy in each nation, correlations between GDP and life expectancy,  and a look at the distributions that show both the GDP and life expectancy data.  

## Importing modules

This project will utilize the following modules: pandas, matplotlib and seaborn.

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

## Loading in data

The data from the all_data.csv will be loaded into a dataframe. 

```python
df = pd.read_csv('all_data.csv')
df.head()
```

Will rename the 'life expectancy at birth (years)' column to 'Life expectancy' as it is longer than other columnn and to allow for more readable visualizations going forward. 

```python
df = df.rename({'Life expectancy at birth (years)':'Life expectancy'}, axis='columns')
df.head()
```

Will create an violin plot to compare distributions with respect to life expectancy.

```python
fig = plt.subplots(figsize=(15, 10))
sns.violinplot(data=df, x='Country', y='Life expectancy')
plt.show()
```

In the diagram above Zimbabwe seems to have a wider range of values in addition to smaller values. The rest of the countries have smaller distributions and life expectancies ranging from 72 to 81 years. 


Seaborn's bar plot will be used on 'Life expectancy' and 'Country' columns.

```python
ax = sns.barplot(data=df, x='Country', y='Life expectancy')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_ylabel('Life Expectancy (years)')
ax.set_title('Average Life Expectancy')
```

## Average GDP By Country

```python
ax = sns.barplot(data=df, y='Country', x='GDP')
ax.set_xlabel('GDP (trillion ($))')
ax.set_title('Average GDP')
```

## Life Expectancy Over Time

```python
sns.lineplot(data=df, x='Year', y='Life expectancy', hue='Country')
plt.ylabel('Life Expectancy (years)')
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
plt.title('Life expectancy over time')
```

This plots shows a marked difference between the life expectancy of Zimbabwe and the other five nations.

The general trend in this graph is that the five other nations outside of Zimbabwe are trending upwards but onlu slightly. There is a small decrease in life expectancy in Zimbabwe after 2004 but there is a dramatic increase as the you move into 2006 onwards.

## GDP Over Time

```python
ax = sns.lineplot(data=df, x='Year', y='GDP', hue='Country')
ax.set_ylabel('GDP (trillion dollars)')
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
plt.title('GDP over time')
```

The United States of America is shown to have the largest GDP of all the nations included in the data set. This visualization also reveals a drastic upward trend in GDP growth for China as well. On the other hand, Mexico, Germany, Chile, and Zimbabwe have had fairly small changes in GDP trends.

```python
ax = sns.scatterplot(data=df, y='GDP', x='Life expectancy', hue='Country')
ax.set_ylabel('GDP (trillion dollars)')
ax.set_xlabel('Life Expectancy (years)')
ax.set_title('Life Expectancy vs GDP')
```

```python
g = sns.FacetGrid(df, col='Country', hue='Country', col_wrap=3, sharey=False, sharex=False)
g.map(sns.scatterplot,'Life expectancy','GDP')
g.set_axis_labels('Life Expectancy (years)','GDP (trillion dollars)')
g.fig.suptitle('Relationship of Life Expectancy and GDP', fontsize=14)
g.fig.tight_layout(rect=[0, 0, 1, 1])
```

In general, it can be seen in each of the scatter plots in the chart above that there is a positive correlation between life expectancy and GDP. The data points on the China show a sort of exponentional curve whereas on the U.S. plot they form a  basically straight line. This indicates a strong positive correlation between the two variables.
